```python
import numpy as np
```

```python
def setDist(pos, map_, distance, visited):
    directs=[(0,1), (1,0), (0,-1), (-1,0)]
    p = tuple(pos)
    reachable = []
    dists = []
    for d in directs:
        nextpos = pos + d
        nxtp = tuple(nextpos)
        # if(p[1] == 1):
        #    print('At', p, 'checking', nxtp)
        if visited[nxtp] == 1:
            # if(p[1] == 1):
            #     print('visited')
            continue  # Already visited
        
        # visited[nxtp] += 1
        if (map_[nxtp] - map_[p])<2 and distance[p]+1 < distance[nxtp]:
            #if(p[1] == 1):
            #    print('updating', distance[nxtp], 'to', distance[p]+1)
            distance[nxtp] = distance[p] + 1                
            
    visited[p] = 1  # Mark as visited
    
def loop(map_, distance, visited):
    
    while True:
        
        nv = np.where(visited == 0)
        
        if len(nv[0]) == 0:
            break
            
        mindist = np.min(distances[nv])
        
        w_x, w_y = np.where((visited==0) & (distances == mindist))
    
        setDist(np.array([w_x[0], w_y[0]]), map_, distance, visited)
```

```python
with open('input12.txt', 'r') as inf:
    data=[l.strip() for l in inf]
    # data=[l.strip() for l in inf]
```

```python
mdata = []
for row in data:
    mdata.append([ord(h) for h in row])
    
orig_map_ = np.array(mdata)
visited = np.full_like(orig_map_, 0)
distances = np.full_like(orig_map_, 1000)
map_ = np.pad(orig_map_, 1, 'constant', constant_values=0)
visited = np.pad(visited, 1, 'constant', constant_values=1)
distances = np.pad(distances, 1, 'constant', constant_values=1000)

s_x, s_y = np.where(map_ == ord('S'))
start = np.array([s_x[0], s_y[0]])

end = np.where(map_==ord('E'))
print('Starting at', start)
print('Ending at', end)

map_[tuple(start)] = ord('a')
map_[end] = ord('z') + 1

distances[tuple(start)] = 0

loop(map_, distances, visited)

print('Shortest route', distances[end][0])

print('\n\nPart 2\n\n')

dists = []

start_x, start_y = np.where(map_ == ord('a'))

print(f'Found {len(start_x)} starting points')

i = 1
for sx, sy in zip(start_x, start_y):
    
    if (i % 20) == 0:
        print(i)
    visited = np.full_like(orig_map_, 0)
    distances = np.full_like(orig_map_, 1000)
    visited = np.pad(visited, 1, 'constant', constant_values=1)
    distances = np.pad(distances, 1, 'constant', constant_values=1000)
    
    distances[sx, sy] = 0

    loop(map_, distances, visited)
    
    dists.append(distances[end][0])
    i+=1
    
print('Shortest possible route', min(dists))
```
