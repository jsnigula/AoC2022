```python
def rsplit(data):
    
    if '[' not in data:
        if len(data) == 0:
            return []
        return int(data)

    if data[0] == '[':
        data = data[1:-1]

    if ',' not in data:
        if '[' in data:
            return [rsplit(data)]
        if len(data) == 0:
            return []
        return [int(data)]
    
    splits = []
    n_block = 0
    curstr = ''
    for s in data:
        # print('Parsing', s, n_block)
        if n_block != 0:
            curstr += s
            if s == ']':
                n_block -= 1
            elif s == '[':
                n_block += 1
        elif s == '[':
            n_block += 1
            curstr += s
        elif s == ',':
            splits.append(curstr)
            curstr = ''
        else:
            curstr += s
            
    splits.append(curstr)
    
    return [rsplit(s) for s in splits]
        
def listcmp(l1, l2):
    if isinstance(l1, int) and isinstance(l2, int):
        if l1 < l2:
            return 0
        elif l1 == l2:
            return 1
        else:
            return 2
    
    if isinstance(l1, list) and isinstance(l2, list):
        i = 0
        while i < len(l1):
            if i >= len(l2):
                return 2
            
            r = listcmp(l1[i], l2[i])
            if r == 0:
                return 0
            if r == 2:
                return 2
            i += 1
        if i == len(l2):
            return 1
        if i < len(l2):
            return 0
        
    if isinstance(l1, list) and isinstance(l2, int):
        return listcmp(l1, [l2])
    
    if isinstance(l1, int) and isinstance(l2, list):
        return listcmp([l1], l2)
        
```

```python
class Packet():
    
    def __init__(self, indata):
        
        self.data = rsplit(indata)
        
        if str(self.data) != str(eval(indata)):
            print(self.data)
            print(eval(indata))
            raise Exception()

    def __str__(self):
        return f'[{",".join([str(d) for d in self.data])}]'
        # return f'Packet len {len(self)}, {self.packets}'
        # r = f'Packet with {len(self.packets)} subpackets:'
        # for p in self.packets:
        #     r += f'\n{p}'
        # return r
            
    def __repr__(self):
        return str(self)
    
    def __len__(self):
        return len(self.packets)

    def __lt__(self, other):
        
        return listcmp(self.data, other.data) == 0
    
    def __eq__(self, other):
        return str(self.data) == str(other.data)
```
