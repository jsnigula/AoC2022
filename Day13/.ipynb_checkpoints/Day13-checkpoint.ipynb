```python
def rsplit(data):
    
    if data[0] == '[':
        data = data[1:-1]

    if ',' not in data:
        if len(data) == 0:
            return None
        return int(data)
    
    splits = []
    n_block = 0
    curstr = ''
    for s in data:
        # print('Parsing', s, n_block)
        if n_block != 0:
            curstr += s
            if s == ']':
                n_block -= 1
            elif s == '[':
                n_block += 1
        elif s == '[':
            n_block += 1
            curstr += s
        elif s == ',':
            splits.append(curstr)
            curstr = ''
        else:
            curstr += s
            
    splits.append(curstr)
    
    return [rsplit(s) for s in splits]
        
```

```python

class Packet():
    
    def __init__(self, indata):
        
        self.data = []
        
        # if ',' not in indata:
        #     print('intparsing', indata)
        #     if '[' in indata:
        #         indata = indata[1:-1]
        #     if len(indata) != 0:
        #         self.packets.append(int(indata))
        #     return
        
        # print('Splitting', indata)
        # print('Solit', psplit(indata))
        for pdata in rsplit(indata):
            # print('Processing', pdata)
            if '[' in pdata:
                self.data.append(Packet(pdata))
            else:
                if len(pdata) != 0:
                    # self.packets.append(Packet(pdata))
                    self.packets.append(int(pdata))

    def __str__(self):
        return f'[{",".join([str(p) for p in self.packets])}]'
        # return f'Packet len {len(self)}, {self.packets}'
        # r = f'Packet with {len(self.packets)} subpackets:'
        # for p in self.packets:
        #     r += f'\n{p}'
        # return r
            
    def __repr__(self):
        return str(self)
    
    def __len__(self):
        return len(self.packets)

    def __lt__(self, other):
        
        print(f'{str(self)} < {str(other)}')
        if isinstance(other, int):
            if len(self.packets) == 0:
                print('False')
                return False
        #print('intcomp', self.packets[0], other)
            return self.packets[0] <= other
        # print(self, '<', other)
        
        if len(self) > len(other):
            # print('shorter')
            print('False')
            return False
                
        if len(self) == 0:
            # print('nolen')
            print('True')
            return True

        for i in range(0, len(self)):
            if isinstance(self.packets[i], int):
                # print('amint')
                if not isinstance(other.packets[i], int):
                #     print('onotint')
                    if not self.packets[i] <= other.packets[i].getVal():
                        print('False')
                        return False
                else:
                    if not self.packets[i] <= other.packets[i]:
                #         print(self.packets[i], '<=', other.packets[i])
                        print('False')
                        return False
            elif not self.packets[i] <= other.packets[i]:
                # print('!', self.packets[i], '<=', other.packets[i])
                print('False')
                return False
            
        print('True')
        return True

    def getVal(self):
        if len(self.packets) == 0:
            return -1
        if isinstance(self.packets[0], int):
            return self.packets[0]
        else:
            return self.packets[0].getVal()
```
