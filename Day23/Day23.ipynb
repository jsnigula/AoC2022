```python
elfpos = []
with open('input23.txt', 'r') as inf:
    for y, line in enumerate(inf):
        for x, c in enumerate(line.strip()):
            if c == '#':
                elfpos.append((x, y))
        
moves = [(0,-1), (0,1), (-1,0), (1,0)]
checks = [[(0,-1), (-1,-1), (1,-1)], [(0,1), (-1,1), (1,1)], [(-1,0), (-1,-1), (-1,1)], [(1,0), (1,-1), (1,1)]]

for r in range(10):
    nextpos = defaultdict(lambda: 0)
    canmove = {}
    for e in elfpos:
        # print('Checking', e)
        canmove[e] = -1
        shouldmove = False
        for i in range(4):
            for checkdir in checks[i]:
                if (e[0]+checkdir[0], e[1]+checkdir[1]) in elfpos:
                    shouldmove = True
                    break
            if shouldmove:
                break
                    
        if shouldmove:
            # print('Shouldmove')
            
            for i in range(4):
                free = True
                for checkdir in checks[i]:
                    if (e[0]+checkdir[0], e[1]+checkdir[1]) in elfpos:
                        free = False
                        break
                if free:
                    canmove[e] = i
                    break

            if canmove[e] != -1:
                mdir = canmove[e]
                nextpos[(e[0]+moves[mdir][0],e[1]+moves[mdir][1])] += 1
            
    # print(canmove)
    # print(nextpos)
    # now move
    newpos = []
    n_moving = 0
    for e in elfpos:
        mdir = canmove[e]
        
        if mdir != -1 and nextpos[(e[0]+moves[mdir][0],e[1]+moves[mdir][1])] < 2:
            newpos.append((e[0]+moves[mdir][0],e[1]+moves[mdir][1]))
            n_moving += 1
        else:
            newpos.append(e)
    elfpos = newpos
    
    moves = [moves[1], moves[2], moves[3], moves[0]]
    checks = [checks[1], checks[2], checks[3], checks[0]]
    
    # print('Moving', n_moving)
    # print(elfpos)
    
minx=1e9
maxx=0
miny=1e9
maxy=0

for e in elfpos:
    x, y = e
    minx = min(x, minx)
    maxx = max(x, maxx)
    miny = min(y, miny)
    maxy = max(y, maxy)
    
print('Empty ground:', (maxx-minx+1) * (maxy-miny+1) - len(elfpos))
```
