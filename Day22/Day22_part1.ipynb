```python
import numpy as np
import re
import copy
```

```python
def turn(curdir, t):
    if t == 'R':
        curdir += 1
    else:
        curdir -= 1
        
    if curdir == -1:
        curdir = 3
    if curdir == 4:
        curdir = 0
        
    return curdir
        
def nextedge(data, nextpos, direct):
    
    # print('Nextwrap', nextpos, direct)
    if direct == 3:
        x = nextpos[1]
        y = np.argwhere(data[:,nextpos[1]] != 0)[-1][0]
        # print(x, np.argwhere(data[:,nextpos[1]] != 0), np.argwhere(data[:,nextpos[1]] != 0)[-1][0], '#-1' )
    elif direct == 0:
        x = np.argwhere(data[nextpos[0],:] != 0)[0][0]
        y = nextpos[0]
        # print(np.argwhere(data[nextpos[0],:] != 0),np.argwhere(data[nextpos[0],:] != 0)[0][0] , y, '#0')
    elif direct == 1:
        x = nextpos[1]
        y = np.argwhere(data[:,nextpos[1]] != 0)[0][0]
        # print(x, np.argwhere(data[:,nextpos[1]] != 0), np.argwhere(data[:,nextpos[1]] != 0)[0][0], '#0')
    else:
        x = np.argwhere(data[nextpos[0],:] != 0)[-1][0]
        y = nextpos[0]
        # print(np.argwhere(data[nextpos[0],:] != 0), np.argwhere(data[nextpos[0],:] != 0)[-1][0], y, '#-1')
        
    return np.array([y, x])
```

```python
line_data = []

lines = ''
maxlen = 0
mapdone = False

with open('input22.txt', 'r') as inf:
    for line in inf:
        if line.strip() == '':
            mapdone = True
            continue
        if mapdone:
            path = line.strip()
            continue
        lines += line
        ld = []
        for v in line:
            if v == '.':
                ld.append(1)
            elif v == '#':
                ld.append(2)
            elif v == '\n':
                continue
            else:
                ld.append(0)
        if len(ld) > maxlen:
            maxlen = len(ld)
        line_data.append(ld)

data = np.zeros((len(line_data)+2, maxlen+2))

for i, ld in enumerate(line_data):
    data[i+1, 1:len(ld)+1] = ld
pathmap = copy.copy(data)
    
dists = re.findall('(\d+)', path)
turns = re.findall('([RL])', path)

# print(data)

pos = np.array([1, np.where(data[1,:] != 0)[0][0]])
direct = 0  # 0 righ, 1 down, 2 left, 3 up

dirmap = {3: np.array([-1, 0]),
          0: np.array([0, 1]),
          1: np.array([1, 0]),
          2: np.array([0, -1])}

pathmap[tuple(pos)] = 4

i = 0
while True:
    
    if i >= len(dists):
        break
    # print('Moving', int(l), d)
    for s in range(int(dists[i])):
        nextpos = pos + dirmap[direct]
        # print('Pos', pos, 'Nextpos', nextpos)
        np_t = tuple(nextpos)
        if data[np_t] == 0: # wrapping around
            # print('Wrap')
            nextpos = nextedge(data, nextpos, direct)
            np_t = tuple(nextpos)
        if data[np_t] == 2:
            # print('Wall')
            break # Do not move further
        else:
            pos = nextpos
            pathmap[tuple(pos)] = 4
            
    # print('Now at', pos)
    if i >= len(turns):
        break
    direct = turn(direct, turns[i])
    # print('Facing', direct)
    i+=1
    
print('Password', 1000*pos[0] + 4*pos[1] + direct)
# print(pathmap)
```
