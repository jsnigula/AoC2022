```python
def turn(curdir, t):
    if t == 'R':
        curdir += 1
    else:
        curdir -= 1
        
    if curdir == -1:
        curdir = 3
    if curdir == 4:
        curdir = 0
        
    return curdir
        
def nextedge(cube, nextpos, direct):
    
    f, pos = nextpos
    y, x = pos
    newdir = direct

    # print(f, pos, direct)
    if f == 'B': # 
        if x == 0: #
            newpos = ('3', np.array([51-y, 1]))
            assert direct == 2
            newdir = 0
        elif x == 51: #
            newpos = ('1', np.array([y, 1]))
            assert direct == 0
            newdir = 0
        elif y == 0: # 
            newpos = ('4', np.array([x, 1]))
            assert direct == 3
            newdir = 0
        else: # 
            newpos = ('2', np.array([1, x]))
            assert direct == 1
            newdir = 1
        
    elif f == 'T': # 
        if x == 0: #
            newpos = ('3', np.array([y, 50]))
            assert direct == 2
            newdir = 2
        elif x == 51: #        
            newpos = ('1', np.array([51-y, 50]))
            assert direct == 0
            newdir = 2
        elif y == 0: #
            newpos = ('2', np.array([50, x]))
            assert direct == 3
            newdir = 3
        else: # 
            newpos = ('4', np.array([x, 50]))
            assert direct == 1
            newdir = 2
        
    elif f == '1':
        if x == 0: #
            newpos = ('B', np.array([y, 50]))
            assert direct == 2
            newdir = 2
        elif x == 51: #
            newpos = ('T', np.array([51-y, 50]))
            assert direct == 0
            newdir = 2
        elif y == 0: # 
            newpos = ('4', np.array([50, x]))
            assert direct == 3
            newdir = 3
        else: #
            newpos = ('2', np.array([x, 50]))
            assert direct == 1
            newdir = 2
        
    elif f == '2':
        if x == 0: #
            newpos = ('3', np.array([1, y]))
            assert direct == 2
            newdir = 1
        elif x == 51: #
            newpos = ('1', np.array([50, y]))
            assert direct == 0
            newdir = 3
        elif y == 0: # 
            newpos = ('B', np.array([50, x]))
            assert direct == 3
            newdir = 3
        else: # 
            newpos = ('T', np.array([1, x]))
            assert direct == 1
            newdir = 1
        
    elif f == '3':
        if x == 0: # 
            newpos = ('B', np.array([51-y, 1]))
            assert direct == 2
            newdir = 0
        elif x == 51: #
            newpos = ('T', np.array([y, 1]))
            assert direct == 0
            newdir = 0
        elif y == 0: #
            newpos = ('2', np.array([x, 1]))
            assert direct == 3
            newdir = 0
        else: #
            newpos = ('4', np.array([1, x]))
            assert direct == 1
            newdir = 1
        
    elif f == '4':
        if x == 0: # 
            newpos = ('B', np.array([1,y]))
            assert direct == 2
            newdir = 1
        elif x == 51: #
            newpos = ('T', np.array([50, y]))
            assert direct == 0
            newdir = 3
        elif y == 0: #
            newpos = ('3', np.array([50, x]))
            assert direct == 3
            newdir = 3
        else: #
            newpos = ('1', np.array([1, x]))
            assert direct == 1
            newdir = 1
        
    return newpos, newdir

def getval(cube, pos):
    
    f, p = pos
    return cube[f][tuple(p)]
```
